```python
#TREATING MISSING VALUES.
# By default, missing values are represented with NaN, "not a number".
# WARNING: If your dataset has 0s, 99s, 0r 999s, be sure to either drop or approximate them as you would with missing values, otherwise you may end up with inaccurate results.
```

```python
# Missin Values in Python
import numpy as np
import pandas as pd

from pandas import DataFrame
```

```python
data = {'names': ['David', 'John', 'Jane', 'Smith', 'Emily', 'richard', 'Abraham'],
        'age': [25, 30, 22, 35, 28, np.nan, 40],
        'gender': ['M', 'M', 'F', 'M', 'F', 'M', np.nan],
        'rank': [2, 1, 4, 5, 3, 6, 7]}

ranking_df = DataFrame(data)
ranking_df.iloc[2:5,1]=np.nan
ranking_df.iloc[3:6,3]=np.nan
ranking_df.iloc[3,:]=np.nan
ranking_df
```

```python
#true represents the value that is missing and false represents the value that is not missing.
ranking_df.isnull()
```
