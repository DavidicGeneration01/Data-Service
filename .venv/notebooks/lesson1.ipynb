```python
import numpy as np
import pandas as pd

from pandas import DataFrame
```

```python
numbers_df = pd.DataFrame(np.arange(0,90,3).reshape(10,3), index=['row 1', 'row 2', 'row 3', 'row 4', 'row 5', 'row 6', 'row 7', 'row 8', 'row 9', 'row 10'], columns=['column 1', 'column 2', 'column 3'])
```

```python
numbers_df
```

```python
# Example of simple indexing
numbers_df.iloc[0,1]
```

```python
numbers_df.iloc[0,1]=20
```

```python
numbers_df
```

```python
# fancy indexing
numbers_df.iloc[[1,2,4],[1,2]]
```

```python
# Example of boolean indexing
mask = numbers_df > 30
```

```python
mask
#  Example of Boolean operators replace numbers with value.
```

```python
numbers_df[mask]
```

```python
numbers_df[numbers_df>30]=0
```
