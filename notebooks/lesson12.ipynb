Benefits of Statistical Plots

Statistical plots allow viewers to:
- Identify outliers
- Visualize distributions
- Deduce variable types
- Discover relationships and core relations between variables in a dataset

```python
import numpy as np
import pandas as pd 
from pandas import Series, DataFrame

from pandas.plotting import scatter_matrix

import matplotlib.pyplot as plt
from pylab import rcParams
import seaborn as sns
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 5, 4
sns.set_style('whitegrid')
```

Eyeballing dataset distributions with histogram

```python
address = '../data/mtcars.csv'

cars = pd.read_csv(address)

cars.columns = ['car_name', 'mpg', 'cyl', 'disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']
cars.index = cars['car_name']  

mpg = cars['mpg']

mpg.plot(kind='hist')
```

```python
plt.hist(mpg)
plt.plot()
```

```python
sns.displot(mpg)
```
