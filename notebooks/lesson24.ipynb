Four Beautiful Soup Object Types
1. BeautifulSoup object
2. Tag object
3. NavigableString object
4. Comment object

The Tag Object
- Tag objects correspond to XML and HTML elements in an original document
- You can navigate and reference data using tag attributes

NavigableString
- A NavigableString object is used to add a bit of text within a tag 
- Beautiful Soup uses the NavigableString class as a container for bits of text

The comment Object: The Comment object is a type of NavigableString object that you can use for commenting your code
