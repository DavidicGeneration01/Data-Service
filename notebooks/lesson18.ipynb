Outlier Detection
- Most machine learning methods assume that your data has been treated for outliers
- Detecting outlier can be a data preprocessing task or an analytical method on its own merits

Outlier Analysis Use Cases:
* Use outlier detection to uncover anomalies that represent:
- To detect Equipment failure
- To detect Fraud
- To detect Cybersecurity events

Interquartile Range (IQR)
Interquartile range (IQR), or spread - the distance between the first quartile (at 25%) and the third quartile (at 75%)

Univariate Method: Tukey Outlier Labeling
a = Q1 - 1.5(IQR)
b = Q3 + 1.5(IQR)
if...
    minimum value<a or
    maximum value > b, then...
The variable is suspect for outliers.
