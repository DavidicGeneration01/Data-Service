```python
url = 'https://raw.githubusercontent.com/selva86/datasets/master/mtcars.csv'
cars = pd.read_csv(url, header=0)

# Rename all 14 columns
cars.columns = ['cars', 'car_names', 'mpg', 'cyl', 'disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb', 'fast']
mpg = cars['mpg']
```

```python
mpg.plot()
```

```python
df = cars[['cyl', 'wt', 'mpg']]
df.plot()
```
