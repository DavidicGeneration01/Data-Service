Creating labels and Annotations


Labels and Annotations Overview
- Labeling plot features
- Adding a legend to your plot
- Annotating your plot


```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame

import matplotlib.pyplot as plt
from pylab import rcParams
import seaborn as sns

```

```python
%matplotlib inline
rcParams['figure.figsize'] = 12, 4
sns.set_style('whitegrid')
```

Labeling plot features


The functional method

```python
x = range(1, 10)
y = [1,2,3,4,0.5,4,3,2,1]

plt.bar(x,y)

plt.xlabel('your x axis label')
plt.ylabel('your y axis label')
```

```python
z= [1,2,3,4,0.5]
veh_type = ['bicycle', 'motorbike', 'car', 'van', 'stroller']
plt.pie(z, labels=veh_type)
plt.show()
```

The object-oriented method


```python
address = 'https://raw.githubusercontent.com/selva86/datasets/master/mtcars.csv'

cars = pd.read_csv(address)
cars.columns = ['mpg', 'cyl', 'disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb', 'fast', 'cars', 'car_names']


mpg = cars.mpg

fig = plt.figure()
ax = fig.add_axes([.1, .1, 1, 1])

mpg.plot()

ax.set_xticks(range(32))

ax.set_xticklabels(cars.car_names, rotation=60, fontsize='medium')
ax.set_title('Miles Per Gallon of Cars in mtcars')

ax.set_xlabel('car names')
ax.set_ylabel('miles/gallon')
```

Adding a legend to your plot
- The functional method


```python
plt.pie(z)
plt.legend(veh_type, loc='best')
plt.show()
```

The object-oriented method

```python
fig = plt.figure()
ax = fig.add_axes([.1, .1, 1, 1])

mpg.plot()

ax.set_xticks(range(32))

ax.set_xticklabels(cars.car_names, rotation=60, fontsize='medium')
ax.set_title('Miles Per Gallon of Cars in mtcars')

ax.set_xlabel('car names')
ax.set_ylabel('miles/gal')

ax.legend(loc='best')
```

Annotating your plot

```python
fig = plt.figure()
ax = fig.add_axes([.1, .1, 1, 1])

mpg.plot()

ax.set_xticks(range(32))

ax.set_xticklabels(cars.car_names, rotation=60, fontsize='medium')
ax.set_title('Miles Per Gallon of Cars in mtcars')

ax.set_xlabel('car names')
ax.set_ylabel('miles/gallon')

ax.set_ylim([0, 45])


#Adding annotation
ax.annotate('Toyota Corolla', xy=(19,33.9), xytext=(21,35), arrowprops=dict(facecolor='black', shrink=0.05))

ax.legend(loc='best')
```
