```python
#GROUPING AND AGGREGATION
# To compare subsets
# To deduce reasons why subgroup differ
# To subset your data for your analysis
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
url = 'https://raw.githubusercontent.com/selva86/datasets/master/mtcars.csv'
cars = pd.read_csv(url, header=0)

# Rename all 14 columns
cars.columns = ['mpg', 'cyl', 'disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb', 'fast', 'cars', 'car_names']

cars.head()
```

```python
cars_groups = cars.groupby(cars['cyl'])
cars_groups.mean(numeric_only=True)
```
