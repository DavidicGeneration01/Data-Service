```python
#TREATING MISSING VALUES.
# By default, missing values are represented with NaN, "not a number".
# WARNING: If your dataset has 0s, 99s, 0r 999s, be sure to either drop or approximate them as you would with missing values, otherwise you may end up with inaccurate results.
```

```python
# Missin Values in Python
import numpy as np
import pandas as pd

from pandas import DataFrame
```

```python
data = {'names': ['David', 'John', 'Jane', 'Smith', 'Emily', 'richard', 'Abraham'],
        'age': [25, 30, 22, 35, 28, np.nan, 40],
        'gender': ['Male', 'Male', 'Female', 'Male', 'Female', 'Male', np.nan],
        'rank': [2, 1, 4, 5, 3, 6, 7]}

ranking_df = DataFrame(data)
ranking_df.iloc[2:5,1]=np.nan
ranking_df.iloc[3:6,3]=np.nan
ranking_df.iloc[3,:]=np.nan
ranking_df
```

```python
#true represents the value that is missing and false represents the value that is not missing.
ranking_df.isnull()
```

```python
ranking_df.notnull()
```

```python
# This is to find the missing values in the age column.
bool_series = pd.isnull(ranking_df['age'])
ranking_df[bool_series]
```

```python
# This function fill missing values in a DataFrame with a specified value with a single value
ranking_df.fillna(0)
```

```python
ranking_df.ffill()
```

```python
ranking_df.bfill()
```

```python
ranking_df = ranking_df.apply(lambda col: col.interpolate(method='linear') if col.dtype in ['float64', 'int64'] else col)
```

```python
# Convert back to float first, then format
ranking_df['age'] = ranking_df['age'].astype(float).map('{:.2f}'.format)
ranking_df['rank'] = ranking_df['rank'].astype(float).map('{:.2f}'.format)

ranking_df
```

```python
ranking_df.dropna()
```

```python
ranking_df.dropna(how='all')
```

```python
ranking_df.dropna(axis=1)
```

```python
ranking_df.dropna(axis=0)
```
