Summarizing Categorical Variable

Categorical Variable
- Accepts only a limited and fixed number of values
- Each observation is assigned to a specific subgroup


Summarizing categorical data using pandas

```python
import numpy as np
import pandas as pd
```

```python
address = '../data/mtcars.csv'

cars = pd.read_csv(address)
cars.columns = ['car_name', 'mpg', 'cyl', 'disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']
cars.index = cars.car_name

cars.head(15)
```

```python
carb = cars.carb
carb.value_counts()
```

```python
cars_cat = cars[['cyl', 'vs', 'am', 'gear', 'carb']]
cars_cat.head()
```

```python
gears_group = cars_cat.groupby('gear')
gears_group.describe()
```

Transforming variables to categorical data type

```python
cars['group'] = pd.Series(cars.gear, dtype='category')
```

```python
cars['group'].dtypes
```

```python
cars['group'].value_counts()
```

Describing categorical data with crosstabs

```python
pd.crosstab(cars['am'], cars['gear'])
```
