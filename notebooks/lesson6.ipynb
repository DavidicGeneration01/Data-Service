```python
import matplotlib.pyplot as plt
import seaborn as sns

from pandas import DataFrame
```

```python
data = {'names': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'john', 'sarah'],
        'ages': [20, 22, 20, 21, 24, 23, 22],
        'gender':['Female', 'Male', 'Male', 'Male', 'Female', 'Male', 'Female'],
        'rank':[2, 1, 4, 5, 3, 7, 6]}

df=DataFrame(data)
df
```

```python
#Matplotlib's Bar Chart
plt.bar(df['names'], df['ages'])
# set label for the axises
plt.xlabel('Names')
plt.ylabel('Ages')
# create a title for the plot
plt.title('Comparing Ages')
plt.show()
```

```python
# plot the same graph using seaborn
plot = sns.barplot(data=df, x='names', y='ages')
# create title for the plot
plot.set_title('Comparing Ages')
plt.show()

#seaborn is more efficient and has more features than Matplotlib, it is built on top of Matplotlib and provides a higher-level interface for creating attractive and informative statistical graphics. Seaborn also has built-in support for visualizing data with categorical variables, which can be useful for comparing groups of data. Additionally, Seaborn has a more modern and aesthetically pleasing default style compared to Matplotlib, which can make it easier to create visually appealing plots.
# And for beautiful plots, seaborn is the best choice.
```

```python
#Line Plot Matplotlib
plt.plot(df['names'], df['ages'])
# set label for the axises
plt.xlabel('Names')
plt.ylabel('Ages')
# create a title for the plot
plt.title('Comparing Ages')
plt.show()
```

```python
#Line Plot Seaborn
plot = sns.lineplot(data=df, x='names', y='ages')
plt.show()
```

```python
#Pie Chart Matplotlib
plt.pie(df['ages'], labels=df['names'])
plt.title("Age Comparison")
plt.show()
```

```python
#Pie Chart Seaborn
colors = sns.color_palette('pastel')[0:5]
plt.pie(df['ages'], labels=df['names'], colors=colors)
plt.show()
```
