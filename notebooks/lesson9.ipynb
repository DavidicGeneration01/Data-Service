Plot Formatting Overview
- Defining plot color
- Customizing line styles
- Setting marker styles

```python
import numpy as np
import pandas as pd
from pandas import DataFrame

import matplotlib.pyplot as plt
from pylab import rcParams

import seaborn as sb
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 5, 4
sb.set_style('whitegrid')
```

Defining Plot Color

```python
x = range(1, 10)
y = [1,2,3,4,0.5,4,3,2,1]

plt.bar(x, y)
```

```python
wide = [0.5,0.5,0.5,0.9,0.9,0.9,0.5,0.5,0.5]
color = ['salmon']

plt.bar(x,y, width=wide, color=color, align='center')
```

```python
address = 'https://raw.githubusercontent.com/selva86/datasets/master/mtcars.csv'
cars = pd.read_csv(address, header=0)
cars.columns = ['mpg', 'cyl', 'disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb', 'fast', 'cars', 'car_names']
mpg = cars['mpg']
```

```python
df = cars[['cyl', 'mpg', 'wt']]
df.plot()
```

```python
#changing default color to your choice color
color_theme = ['darkgray', 'lightsalmon', 'powderblue']
df.plot(color=color_theme)
```

```python
z= [1,2,3,4,0.5]
plt.pie(z)
plt.show()
```

```python
#changing default color to your choice color
color_theme = ['#a9a9a9', '#ffa07a', '#b0e0e6', '#fff4c4', '#bdb76b']
plt.pie(z, colors=color_theme)
plt.show()
```

Customizing line styles

```python
x1 = range(1,10)
y1 = [9,8,7,6,5,4,3,2,1]
plt.plot(x, y)
plt.plot(x1, y1)
```

```python
plt.plot(x, y, ls = 'solid', lw = 5)
plt.plot(x1, y1, ls='--', lw = 10)
```

Setting plot markers

```python
plt.plot(x, y, marker = '1', mew=20)
plt.plot(x1, y1, marker='+', mew=15)
```
